##   Create Diversion Dataframe

<a id="simplify"></a>

1. Load preprocessed data. 
2. Remove rows for only canceled flights.
3. Remove columns for status, causes of delays and cancellation and weather for simplicity.
4. Add new column "Diversion Label".

````python
#=======================================================================
# Import as DataFrames
#=======================================================================

df_utc = pd.read_pickle('../data/preprocessed_data/df_utc_weather.pkl')
df_utc_durations_2 = pd.read_pickle('../data/preprocessed_data/df_utc_durations_2.pkl')

#=======================================================================
# Simplify DataFrames
#=======================================================================

def del_can_causes(df, df_name):
    # To be kept or removed after import (before the synthetic generation) for simplicity:
    simplify = ["Status", "Cancellation Code", 
                "Carrier Delay (min)", "Weather Delay (min)", "National Air System Delay (min)", 
                "Security Delay (min)", "Late Aircraft Delay (min)",

                "Origin_temperature_C", "Origin_dew_point_C",
                "Origin_wind_speed_KMH", "Origin_wind_direction_deg",
                "Origin_visibility_M", "Origin_pressure_hPa",
                "Origin_weather", "Origin_cloud_cover",
                "Destination_temperature_C", "Destination_dew_point_C", 
                "Destination_wind_speed_KMH", "Destination_wind_direction_deg",
                "Destination_visibility_M", "Destination_pressure_hPa",
                "Destination_weather", "Destination_cloud_cover"]

    # Delete rows for cancelled flights
    print("=" * 75 + "\n" + color.BOLD + color.PURPLE + df_name + color.END)
    print("Shape before --> ", df.shape)
    n_cancelled = df[df["Status"] == "C"].shape[0]
    print("-"*75, "\nDeleting {} cancelled flights.".format(n_cancelled))

    df.drop(df[df["Status"] == "C"].index, inplace=True)
    df.reset_index(drop=True, inplace=True)

    # Adding new column "Diversion Label"
    df["Diversion Label"] = np.where(df["Status"] == "D", 1, 0)
    
    # Delete columns of status, causes of delays and cancellation and weather for simplicity
    print("-"*75, "\nDeleting {} columns {} for simplicity.".format(len(simplify), simplify))
    df.drop(simplify, axis=1, inplace=True)

    print("-"*75)
    print("Shape after --> ", df.shape)
    print("="*75)
    print("")
    return df

# Remove canceled flight, causes of delays and cancellation and weather for simplicity
# Add "Diversion Label"
df_utc = del_can_causes(df_utc, "df_utc")
df_utc_durations_2 = del_can_causes(df_utc_durations_2,
                                        "df_utc_durations_2")

#=======================================================================
# Diversions dataframe
#=======================================================================
df_utc_durations_2_diverted  = df_utc_durations_2[df_utc_durations_2["Diversion Label"]==1]
df_utc_durations_2_diverted.reset_index(drop=True, inplace=True)
````

### Recalculate removed features in ``df_utc_durations_2``
- <font color="red">Adding "Diversion Label" to the `sortred_features`</font>

````python
sorted_features = ["Unique Carrier Code", "Tail Number", 
                    "Origin Airport ID", "ICAO Origin Airport", "Origin City", "Origin State Code", "Origin State Name", 
                    "Destination Airport ID", "ICAO Destination Airport", "Destination City", "Destination State Code", "Destination State Name", 
                    "Quarter", "Day of Week", 
                    "Scheduled Departure Time UTC", "Actual Departure Time UTC", "Departure ΔT (min)", "Departure Delay Label", 
                    "Taxi Out Time (min)", "Wheels Off Time UTC", 
                    "Wheels On Time UTC", "Taxi In Time (min)", 
                    "Scheduled Arrival Time UTC", "Actual Arrival Time UTC", "Arrival ΔT (min)", "Arrival Delay Label", 
                    "Scheduled Elapsed Time (min)", "Actual Elapsed Time (min)", "Air Time (min)", "Distance (miles)", 
                    "Diversion Label",

                    "Status", "Cancellation Code", 
                    "Carrier Delay (min)", "Weather Delay (min)", "National Air System Delay (min)", 
                    "Security Delay (min)", "Late Aircraft Delay (min)",
                    
                    "Origin_temperature_C", "Origin_dew_point_C",
                    "Origin_wind_speed_KMH", "Origin_wind_direction_deg",
                    "Origin_visibility_M", "Origin_pressure_hPa",
                    "Origin_weather", "Origin_cloud_cover",
                    "Destination_temperature_C", "Destination_dew_point_C", 
                    "Destination_wind_speed_KMH", "Destination_wind_direction_deg",
                    "Destination_visibility_M", "Destination_pressure_hPa",
                    "Destination_weather", "Destination_cloud_cover"]
````

**Real data**

````python
#========================("Quarter", "Day of Week")=========================
# Create "Quarter" and "Day of Week" (Monday=1, Sunday=7)
real_flight_information["Quarter"] = real_flight_information["Scheduled Departure Time UTC"].dt.quarter
real_flight_information["Day of Week"] = real_flight_information["Scheduled Departure Time UTC"].dt.weekday + 1

#===========================("Distance (miles)")============================
# Check if the distance is unique for each unique pair of airport IDs
is_unique_distance = (
    df_utc.groupby(["Origin Airport ID", "Destination Airport ID"])["Distance (miles)"]
    .nunique().eq(1).all())

if is_unique_distance:
    print("The distance is unique for each unique pair of airport IDs.\n")
else:
    print("There are pairs of airport IDs with multiple distances.\n")

# Create a DataFrame with unique pairs of airport IDs and corresponding distances
df_distance = df_utc.drop_duplicates(subset=["Origin Airport ID", "Destination Airport ID", "Distance (miles)"])[
    ["Origin Airport ID", "Destination Airport ID", "Distance (miles)"]
].reset_index(drop=True)

# Merge real_flight_information DataFrame with df_distance to add the "Distance (miles)" column
real_flight_information = real_flight_information.merge(df_distance, on=["Origin Airport ID", "Destination Airport ID"], how="left")

#===================(ICAO, City, State Code, State Name)====================
# Collect unique airport IDs with their information
origin_info = df_utc[[
    "Origin Airport ID",
    "ICAO Origin Airport",
    "Origin City",
    "Origin State Code",
    "Origin State Name"
]].rename(columns={
    "Origin Airport ID": "Airport ID",
    "ICAO Origin Airport": "ICAO Airport",
    "Origin City": "City",
    "Origin State Code": "State Code",
    "Origin State Name": "State Name"
})

destination_info = df_utc[[
    "Destination Airport ID",
    "ICAO Destination Airport",
    "Destination City",
    "Destination State Code",
    "Destination State Name"
]].rename(columns={
    "Destination Airport ID": "Airport ID",
    "ICAO Destination Airport": "ICAO Airport",
    "Destination City": "City",
    "Destination State Code": "State Code",
    "Destination State Name": "State Name"
})

# Combine origin and destination information
airports_info = pd.concat([origin_info, destination_info], axis=0).drop_duplicates(subset=["Airport ID"]).reset_index(drop=True)

# Check the consistancy with the number of airports
print("Number of unique origin airports: ", df_utc["Origin Airport ID"].nunique())
print("Number of unique destination airports: ", df_utc["Destination Airport ID"].nunique())
print("Number of airports in 'airports_info' dataframe: ", airports_info.shape[0])
print("")

# Merge origin information
real_flight_information = real_flight_information.merge(
    airports_info.rename(columns={
        "Airport ID": "Origin Airport ID",
        "ICAO Airport": "ICAO Origin Airport",
        "City": "Origin City",
        "State Code": "Origin State Code",
        "State Name": "Origin State Name"
    }),
    on="Origin Airport ID",
    how="left"
)

# Merge destination information
real_flight_information = real_flight_information.merge(
    airports_info.rename(columns={
        "Airport ID": "Destination Airport ID",
        "ICAO Airport": "ICAO Destination Airport",
        "City": "Destination City",
        "State Code": "Destination State Code",
        "State Name": "Destination State Name"
    }),
    on="Destination Airport ID",
    how="left"
)

#==============================(Time features)==============================
# 1. Departure Delay Label (NaN when "Departure ΔT (min)" = NaN)
real_flight_information["Departure Delay Label"] = real_flight_information["Departure ΔT (min)"].apply(lambda x: 1.0 if pd.notna(x) and x >= 15 else (0.0 if pd.notna(x) else np.nan))

# 2. Wheels Off Time UTC
real_flight_information["Wheels Off Time UTC"] = real_flight_information["Actual Departure Time UTC"] + pd.to_timedelta(real_flight_information["Taxi Out Time (min)"], unit="m")

# 3. Wheels On Time UTC
real_flight_information["Wheels On Time UTC"] = real_flight_information["Wheels Off Time UTC"] + pd.to_timedelta(real_flight_information["Air Time (min)"], unit="m")

# 4. Scheduled Arrival Time UTC
real_flight_information["Scheduled Arrival Time UTC"] = real_flight_information["Scheduled Departure Time UTC"] + pd.to_timedelta(real_flight_information["Scheduled Elapsed Time (min)"], unit="m")

# 5. Actual Arrival Time UTC
real_flight_information["Actual Arrival Time UTC"] = real_flight_information["Actual Departure Time UTC"] + pd.to_timedelta(real_flight_information["Actual Elapsed Time (min)"], unit="m")

# 6. Arrival Delay Label (NaN when "Arrival ΔT (min)" = NaN)
real_flight_information["Arrival Delay Label"] = real_flight_information["Arrival ΔT (min)"].apply(lambda x: 1.0 if pd.notna(x) and x >= 15 else (0.0 if pd.notna(x) else np.nan))

#=============================(Sorting columns)=============================
real_flight_information = real_flight_information[sorted_features[:-23]]
````

````python
#========================("Quarter", "Day of Week")=========================
# Create "Quarter" and "Day of Week" (Monday=1, Sunday=7)
synthetic_flight_information["Quarter"] = synthetic_flight_information["Scheduled Departure Time UTC"].dt.quarter
synthetic_flight_information["Day of Week"] = synthetic_flight_information["Scheduled Departure Time UTC"].dt.weekday + 1

#===========================("Distance (miles)")============================
# Check if the distance is unique for each unique pair of airport IDs
is_unique_distance = (
    df_utc.groupby(["Origin Airport ID", "Destination Airport ID"])["Distance (miles)"]
    .nunique().eq(1).all())

if is_unique_distance:
    print("The distance is unique for each unique pair of airport IDs.\n")
else:
    print("There are pairs of airport IDs with multiple distances.\n")

# Create a DataFrame with unique pairs of airport IDs and corresponding distances
df_distance = df_utc.drop_duplicates(subset=["Origin Airport ID", "Destination Airport ID", "Distance (miles)"])[
    ["Origin Airport ID", "Destination Airport ID", "Distance (miles)"]
].reset_index(drop=True)

# Merge synthetic_flight_information DataFrame with df_distance to add the "Distance (miles)" column
synthetic_flight_information = synthetic_flight_information.merge(df_distance, on=["Origin Airport ID", "Destination Airport ID"], how="left")

#===================(ICAO, City, State Code, State Name)====================
# Collect unique airport IDs with their information
origin_info = df_utc[[
    "Origin Airport ID",
    "ICAO Origin Airport",
    "Origin City",
    "Origin State Code",
    "Origin State Name"
]].rename(columns={
    "Origin Airport ID": "Airport ID",
    "ICAO Origin Airport": "ICAO Airport",
    "Origin City": "City",
    "Origin State Code": "State Code",
    "Origin State Name": "State Name"
})

destination_info = df_utc[[
    "Destination Airport ID",
    "ICAO Destination Airport",
    "Destination City",
    "Destination State Code",
    "Destination State Name"
]].rename(columns={
    "Destination Airport ID": "Airport ID",
    "ICAO Destination Airport": "ICAO Airport",
    "Destination City": "City",
    "Destination State Code": "State Code",
    "Destination State Name": "State Name"
})

# Combine origin and destination information
airports_info = pd.concat([origin_info, destination_info], axis=0).drop_duplicates(subset=["Airport ID"]).reset_index(drop=True)

# Check the consistancy with the number of airports
print("Number of unique origin airports: ", df_utc["Origin Airport ID"].nunique())
print("Number of unique destination airports: ", df_utc["Destination Airport ID"].nunique())
print("Number of airports in 'airports_info' dataframe: ", airports_info.shape[0])
print("")

# Merge origin information
synthetic_flight_information = synthetic_flight_information.merge(
    airports_info.rename(columns={
        "Airport ID": "Origin Airport ID",
        "ICAO Airport": "ICAO Origin Airport",
        "City": "Origin City",
        "State Code": "Origin State Code",
        "State Name": "Origin State Name"
    }),
    on="Origin Airport ID",
    how="left"
)

# Merge destination information
synthetic_flight_information = synthetic_flight_information.merge(
    airports_info.rename(columns={
        "Airport ID": "Destination Airport ID",
        "ICAO Airport": "ICAO Destination Airport",
        "City": "Destination City",
        "State Code": "Destination State Code",
        "State Name": "Destination State Name"
    }),
    on="Destination Airport ID",
    how="left"
)

#==============================(Time features)==============================
# 1. Departure Delay Label (NaN when "Departure ΔT (min)" = NaN)
synthetic_flight_information["Departure Delay Label"] = synthetic_flight_information["Departure ΔT (min)"].apply(lambda x: 1.0 if pd.notna(x) and x >= 15 else (0.0 if pd.notna(x) else np.nan))

# 2. Wheels Off Time UTC
synthetic_flight_information["Wheels Off Time UTC"] = synthetic_flight_information["Actual Departure Time UTC"] + pd.to_timedelta(synthetic_flight_information["Taxi Out Time (min)"], unit="m")

# 3. Wheels On Time UTC
synthetic_flight_information["Wheels On Time UTC"] = synthetic_flight_information["Wheels Off Time UTC"] + pd.to_timedelta(synthetic_flight_information["Air Time (min)"], unit="m")

# 4. Scheduled Arrival Time UTC
synthetic_flight_information["Scheduled Arrival Time UTC"] = synthetic_flight_information["Scheduled Departure Time UTC"] + pd.to_timedelta(synthetic_flight_information["Scheduled Elapsed Time (min)"], unit="m")

# 5. Actual Arrival Time UTC
synthetic_flight_information["Actual Arrival Time UTC"] = synthetic_flight_information["Actual Departure Time UTC"] + pd.to_timedelta(synthetic_flight_information["Actual Elapsed Time (min)"], unit="m")

# 6. Arrival Delay Label (NaN when "Arrival ΔT (min)" = NaN)
synthetic_flight_information["Arrival Delay Label"] = synthetic_flight_information["Arrival ΔT (min)"].apply(lambda x: 1.0 if pd.notna(x) and x >= 15 else (0.0 if pd.notna(x) else np.nan))

#=============================(Sorting columns)=============================
synthetic_flight_information = synthetic_flight_information[sorted_features[:-23]]
````

````python
def scale_real_synth(real, synthetic):
    # To apply scalling to the numerical columns only
    numeric_cols = real.select_dtypes(include=['number']).columns
    print("Scaling numeric columns: ", numeric_cols.tolist())

    # Replace NaN with 0 to avoide problems with PCA
    real_numeric = real[numeric_cols].fillna(0, inplace=False)
    synthetic_numeric = synthetic[numeric_cols].fillna(0, inplace=False)

    scaler = MinMaxScaler(
    )  # I am dealing with times and +ve values so [0,1] is fine
    # scaler = StandardScaler() # Bad as the mean will be around 0

    # Fit to real only
    scaled_real = pd.DataFrame(scaler.fit_transform(real_numeric),
                               columns=real_numeric.columns,
                               index=real_numeric.index)

    scaled_synthetic = pd.DataFrame(scaler.transform(synthetic_numeric),
                                    columns=synthetic_numeric.columns,
                                    index=synthetic_numeric.index)
    return scaled_real, scaled_synthetic


#===========================================================================
def check_pca_explained_variance(scaled_real):
    plt.rcdefaults()
    pca = PCA()
    real_flight_information_PCA = pca.fit_transform(scaled_real)
    real_flight_information_PCA = pd.DataFrame(real_flight_information_PCA,
                                               index=scaled_real.index)

    p_components = range(1, pca.n_components_ + 1)
    plt.figure(figsize=(15, 5))
    plt.rc('font', size=15)
    plt.bar(p_components,
            pca.explained_variance_ratio_ * 100,
            color="blue",
            label='Individual explained variance')
    plt.plot(p_components,
             np.cumsum(pca.explained_variance_ratio_ * 100),
             marker='o',
             color="red",
             label='Cumulative explained variance')
    plt.xlabel('Principal components')
    plt.ylabel('Explained variance ratio (%)')
    plt.xticks(p_components)
    plt.legend(loc='best')
    plt.title("Importance of the Principal Components")
    plt.show()


#===========================================================================
def check_diversity_pca(scaled_real, scaled_synthetic, plot_real_on_top=False):
    warnings.simplefilter("ignore", UserWarning)
    plt.rcdefaults()
    pca = PCA(n_components=2, svd_solver='auto')
    # Fit to real only
    real_flight_information_PCA = pca.fit_transform(scaled_real)
    real_flight_information_PCA = pd.DataFrame(real_flight_information_PCA,
                                               index=scaled_real.index)
    # For synthetic data
    synthetic_flight_information_PCA = pca.transform(scaled_synthetic)
    synthetic_flight_information_PCA = pd.DataFrame(
        synthetic_flight_information_PCA, index=scaled_synthetic.index)

    # Scatter plot of the first two principal components
    plt.figure(figsize=(4, 5))

    if plot_real_on_top:
        plt.scatter(synthetic_flight_information_PCA[0],
                    synthetic_flight_information_PCA[1],
                    s=10,
                    color='red',
                    alpha=0.6,
                    label='Syn. Flight Diversions')
        plt.scatter(real_flight_information_PCA[0],
                    real_flight_information_PCA[1],
                    s=10,
                    color='blue',
                    alpha=0.6,
                    label='Real Flight Diversions')
    else:
        plt.scatter(real_flight_information_PCA[0],
                    real_flight_information_PCA[1],
                    s=10,
                    color='blue',
                    alpha=0.6,
                    label='Real Flight Diversions')
        plt.scatter(synthetic_flight_information_PCA[0],
                    synthetic_flight_information_PCA[1],
                    s=10,
                    color='red',
                    alpha=0.6,
                    label='Syn. Flight Diversions')
    plt.xlabel('$1^{{st}}$ PC ({}% variance)'.format(
        round(pca.explained_variance_ratio_[0] * 100, 1)),
               fontsize=14)
    plt.ylabel('$2^{{nd}}$ PC ({}% variance)'.format(
        round(pca.explained_variance_ratio_[1] * 100, 1)),
               fontsize=14)
    plt.title('PCA: Real vs Syn. Flight Diversions')
    plt.legend()
    plt.grid(True)
    plt.show()


#===========================================================================
def check_class_balance(flight_info):
    # Get the id of the object passed to compare
    object_id = id(flight_info)
    # Find the variable name dynamically using globals()
    var_name = [
        name for name, value in globals().items() if id(value) == object_id
    ][0]
    var_name = var_name.split("_")[0].capitalize()
    var_name = "Syn." if var_name == "Synthetic" else var_name

    plt.rcdefaults()
    print("=" * 75)

    no_flights = flight_info.shape[0]
    delayed_dep = flight_info["Departure Delay Label"].value_counts()[1]
    on_time_dep = flight_info["Departure Delay Label"].value_counts()[0]
    delayed_arr = flight_info["Arrival Delay Label"].value_counts()[1]
    #     For diverted flights, no on-time arrivals
    #     on_time_arr = flight_info["Arrival Delay Label"].value_counts()[0]
    on_time_arr = 0

    print("Total number of flights --> ", no_flights)
    print("Departures --> {} on time, {} delayed, {} unknown".format(
        on_time_dep, delayed_dep, no_flights - on_time_dep - delayed_dep))
    print("Arrivals --> {} on time, {} delayed, {} unknown".format(
        on_time_arr, delayed_arr, no_flights - on_time_arr - delayed_arr))
    print("=" * 75)

    features = ["Departure Delay Label", "Arrival Delay Label"]
    fig, axs = plt.subplots(1, 2, figsize=(5, 3))
    axs = axs.flatten()

    # Define color mapping
    colors = {
        '1.0': '#ff7f0e',  # Orange for 1
        '0.0': '#1f77b4'  # Blue for 0
    }

    for i, feature in enumerate(features):
        value_counts = flight_info[feature].value_counts(dropna=False)

        # Sort the values to ensure consistent ordering (0 first, then 1)
        value_counts = value_counts.sort_index()

        labels = value_counts.index.astype(str)
        sizes = value_counts.values

        # Map colors to the corresponding values
        color_list = [colors[label] for label in labels]

        # Define function for percentage formatting with larger font
        def make_autopct(values):
            def my_autopct(pct):
                return f'{pct:.1f}%'

            return my_autopct

        # Increased font sizes for labels and percentage values
        axs[i].pie(sizes,
                   labels=labels,
                   colors=color_list,
                   autopct=make_autopct(sizes),
                   startangle=140,
                   textprops={'fontsize': 14
                              })  # This applies to both labels and percentages
        axs[i].axis('equal')
        axs[i].set_title("{} {}".format(var_name, feature), fontsize=12)

    plt.tight_layout()
    plt.show()
````

- Column shapes = statistical similarity between rows in synthetic vs real data.

| Column Type  | Metric        |
|:-------------|:--------------|
| numerical    | KSComplement  |
| datetime     | KSComplement  |
| boolean      | TVComplement  |
| categorical  | TVComplement  |


- The ``KSComplement`` uses the Kolmogorov-Smirnov statistic. To compute this statistic, we convert a numerical distribution into its cumulative distribution function (CDF). The KS statistic is the maximum difference between the two CDFs, as shown below. [Link](https://docs.sdv.dev/sdmetrics/metrics/metrics-glossary/kscomplement)

- The ``TVComplement`` computes the Total Variation Distance (TVD) between the real and synthetic columns. To do this, it first computes the frequency of each category value and expresses it as a probability. [Link](https://docs.sdv.dev/sdmetrics/metrics/metrics-glossary/tvcomplement)

## <font color = 'Purple'>Fidelity Assessment (discriminative score)</font>

````python
def print_discriminative_score(results, print_table=True, return_df=True):
    # Prepare header and table data
    header = ["Model", "Average Accuracy", "Average F1 Score"]
    table_data = [[
        model, metrics["Average Accuracy"], metrics["Average F1 Score"]
    ] for model, metrics in results.items()]

    # Convert to DataFrame
    df = pd.DataFrame(table_data, columns=header)
    df.set_index("Model", inplace=True)

    # Calculate averages and add as a new row
    avg_values = df.mean().to_frame().T
    avg_values.index = ["Average"]
    df = pd.concat([df, avg_values])

    # Print table with index starting from 1
    if print_table:
        df_display = df.reset_index()
        # Add a new index column starting from 1
        df_display.index = df_display.index + 1
        print(
            tabulate(df_display,
                     headers=header,
                     tablefmt="grid",
                     floatfmt=".4f",
                     showindex=True))

    # Return DataFrame
    if return_df:
        return df
````

- **Load real and synthetic**

````python
# Load
real_flight_information = pd.read_pickle(
    '../data/uc5_outputs/real_vs_synth/real_GC_durations_2_diverted_300.pkl')
synthetic_flight_information = pd.read_pickle(
    '../data/uc5_outputs/real_vs_synth/synthetic_GC_durations_2_diverted_300.pkl'
)

# df_utc_duration_2_diverted used in the generation
features = [
    "Unique Carrier Code", "Tail Number", "Origin Airport ID",
    "Destination Airport ID", "Scheduled Departure Time UTC",
    "Actual Departure Time UTC", "Departure ΔT (min)", "Taxi Out Time (min)",
    "Taxi In Time (min)", "Arrival ΔT (min)", "Scheduled Elapsed Time (min)",
    "Actual Elapsed Time (min)", "Air Time (min)", "Diversion Label"
]

real_flight_information = real_flight_information[features]
synthetic_flight_information = synthetic_flight_information[features]
````

**Encode non-numeric features:**

````python
# Encode categorical
categorical_columns = [
    "Unique Carrier Code", "Tail Number", "Origin Airport ID",
    "Destination Airport ID"
]
# 1. ordinal encoding (to handel unseen categories)
ordinal_encoder = OrdinalEncoder(handle_unknown='use_encoded_value',
                                 unknown_value=-1)
for col in categorical_columns:
    # Fit the encoder on real data (reshape column to 2D)
    real_flight_information[col] = ordinal_encoder.fit_transform(
        real_flight_information[[col]])
    # Transform the synthetic dataset using the same encoding
    synthetic_flight_information[col] = ordinal_encoder.transform(
        synthetic_flight_information[[col]])
#==============================================================================
# Encode datetime (convert timestamps to seconds since the Unix epoch)
datetime_columns = [
    "Scheduled Departure Time UTC", "Actual Departure Time UTC"
]
#                     "Wheels Off Time UTC", "Wheels On Time UTC",
#                     "Scheduled Arrival Time UTC", "Actual Arrival Time UTC"]

for col in datetime_columns:
    real_flight_information[col] = pd.to_datetime(
        real_flight_information[col]).dt.tz_convert(None).astype(
            'int64') // 10**9
    synthetic_flight_information[col] = pd.to_datetime(
        synthetic_flight_information[col]).dt.tz_convert(None).astype(
            'int64') // 10**9
#==============================================================================
# Check for non_ numeric columns
print(
    "Non-numeric columns in real data: ",
    real_flight_information.select_dtypes(
        exclude=[np.number]).columns.tolist())
print(
    "Non-numeric columns in synthetic data: ",
    synthetic_flight_information.select_dtypes(
        exclude=[np.number]).columns.tolist())
````

**Get discriminative scores (low is better == synthetic is similar to real):**

````python
# Suppress warnings
warnings.filterwarnings("ignore", category=FutureWarning)

# Combine real and synthetic data
real_flight_information['label'] = 1  # Real data labeled as 1
synthetic_flight_information['label'] = 0  # Synthetic data labeled as 0
combined_data = pd.concat(
    [real_flight_information, synthetic_flight_information])

# Features and labels
X = combined_data.drop(columns=['label'])
y = combined_data['label']

# Models to evaluate
models = {
    "Random Forest":
    RandomForestClassifier(random_state=42),
    "K-Nearest Neighbors (KNN)":
    KNeighborsClassifier(n_neighbors=5),
    "Decision Tree":
    DecisionTreeClassifier(random_state=42),
    "Gradient Boosting":
    GradientBoostingClassifier(random_state=42),
    "Naive Bayes":
    GaussianNB(),
    "Logistic Regression":
    LogisticRegression(random_state=42, max_iter=1000),
    "Stochastic Gradient Descent (SGD)":
    SGDClassifier(loss='hinge', random_state=42),
}

# StratifiedKFold to ensure that class distribution is preserved in each fold
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Evaluate each model and store results
results = {}

for model_name, model in models.items():
    accuracy_scores = cross_val_score(model, X, y, cv=cv, scoring='accuracy')
    f1_scores = cross_val_score(model,
                                X,
                                y,
                                cv=cv,
                                scoring=make_scorer(f1_score))

    results[model_name] = {
        "Average Accuracy": accuracy_scores.mean(),
        "Accuracy StdDev": accuracy_scores.std(),
        "Average F1 Score": f1_scores.mean(),
        "F1 Score StdDev": f1_scores.std(),
    }

    print(f"{model_name}:")
    print(
        f"  Average Accuracy: {accuracy_scores.mean():.4f} (+/- {accuracy_scores.std():.4f})"
    )
    print(
        f"  Average F1 Score: {f1_scores.mean():.4f} (+/- {f1_scores.std():.4f})"
    )
    print("-" * 75)

# Save results to a file
with open(
        "../data/uc5_outputs/discriminative_score/discriminative_score_GC_durations_2_diverted_300.pkl",
        "wb") as f:
    pickle.dump(results, f)
````

**Load real data:**

* Remove canceled flight, causes of delays and cancellation and weather for simplicity
* Add `Diversion Label`.

````python
# Load real data (majority arrived flights, minority diverted flights)
df_utc = pd.read_pickle('../data/preprocessed_data/df_utc_weather.pkl')


def del_can_causes(df):
    # To be kept or removed after import (before the synthetic generation) for simplicity:
    simplify = [
        "Status", "Cancellation Code", "Carrier Delay (min)",
        "Weather Delay (min)", "National Air System Delay (min)",
        "Security Delay (min)", "Late Aircraft Delay (min)",
        "Origin_temperature_C", "Origin_dew_point_C", "Origin_wind_speed_KMH",
        "Origin_wind_direction_deg", "Origin_visibility_M",
        "Origin_pressure_hPa", "Origin_weather", "Origin_cloud_cover",
        "Destination_temperature_C", "Destination_dew_point_C",
        "Destination_wind_speed_KMH", "Destination_wind_direction_deg",
        "Destination_visibility_M", "Destination_pressure_hPa",
        "Destination_weather", "Destination_cloud_cover"
    ]

    # Delete rows for cancelled flights
    print("=" * 75)
    print("Shape before --> ", df.shape)
    n_cancelled = df[df["Status"] == "C"].shape[0]
    print("-" * 75, "\nDeleting {} cancelled flights.".format(n_cancelled))

    df.drop(df[df["Status"] == "C"].index, inplace=True)
    df.reset_index(drop=True, inplace=True)

    # Adding new column "Diversion Label"
    df["Diversion Label"] = np.where(df["Status"] == "D", 1, 0)

    # Delete columns of status, causes of delays and cancellation and weather for simplicity
    print(
        "-" * 75, "\nDeleting {} columns {} for simplicity.".format(
            len(simplify), simplify))
    df.drop(simplify, axis=1, inplace=True)

    print("-" * 75)
    print("Shape after --> ", df.shape)
    print("=" * 75)
    print("")
    return df


# Remove canceled flight, causes of delays and cancellation and weather for simplicity
# Add "Diversion Label"
df_utc = del_can_causes(df_utc)
real = df_utc.copy()
````

**Load synthetic data:**

* Load synthetic diversions to be added to the real training data later.

````python
# Number of samples before cleaning
# 300, 500, 1000, 2000, 5000, 10000, 20000
num_samples = 1000

# Load synthetic diveted flights (to be used for augmentation)
synthetic_diversions = pd.read_pickle(
    '../data/uc5_outputs/real_vs_synth/synthetic_GC_durations_2_diverted_{}.pkl'
    .format(num_samples))
````

**Encode non-numeric features:**

````python
# Encode categorical
categorical_columns = [
    "Unique Carrier Code", "Tail Number", "ICAO Origin Airport",
    "ICAO Destination Airport"
]
# 1. ordinal encoding (to handel unseen categories)
ordinal_encoder = OrdinalEncoder(handle_unknown='use_encoded_value',
                                 unknown_value=-1)
for col in categorical_columns:
    # Fit the encoder on real data (reshape column to 2D)
    real[col] = ordinal_encoder.fit_transform(real[[col]])
    # Transform the synthetic_diversions dataset using the same encoding
    synthetic_diversions[col] = ordinal_encoder.transform(
        synthetic_diversions[[col]])
#==============================================================================
# Encode datetime (convert timestamps to seconds since the Unix epoch)
datetime_columns = [
    "Scheduled Departure Time UTC", "Actual Departure Time UTC",
    "Wheels Off Time UTC", "Wheels On Time UTC", "Scheduled Arrival Time UTC",
    "Actual Arrival Time UTC"
]

for col in datetime_columns:
    real[col] = pd.to_datetime(
        real[col]).dt.tz_convert(None).astype('int64') // 10**9
    synthetic_diversions[col] = pd.to_datetime(
        synthetic_diversions[col]).dt.tz_convert(None).astype('int64') // 10**9
#==============================================================================
remove = [
    "Origin Airport ID", "Origin City", "Origin State Code",
    "Origin State Name", "Destination Airport ID", "Destination City",
    "Destination State Code", "Destination State Name",
    "Departure Delay Label", "Wheels On Time UTC", "Taxi In Time (min)",
    "Actual Arrival Time UTC", "Arrival ΔT (min)", "Arrival Delay Label",
    "Actual Elapsed Time (min)", "Air Time (min)"
]

target_column = "Diversion Label"
real = real.drop(columns=remove)
synthetic_diversions = synthetic_diversions.drop(columns=remove)
#==============================================================================
# Check for non_ numeric columns
print("Non-numeric columns in real data: ",
      real.select_dtypes(exclude=[np.number]).columns.tolist())
print("Non-numeric columns in synthetic_diversions data: ",
      synthetic_diversions.select_dtypes(exclude=[np.number]).columns.tolist())
````

**Classification:**

**Experiment_1 (TRTR)**

````python
features = list(real.columns)

X = real[features[:-1]]
Y = real["Diversion Label"]

# Split with stratification
X_train, X_test, Y_train, Y_test = train_test_split(X,
                                                    Y,
                                                    test_size=0.3,
                                                    stratify=Y,
                                                    random_state=42)

# Define models
models = {
    "Logistic Regression":
    LogisticRegression(class_weight="balanced", max_iter=1000),
    "Random Forest":
    RandomForestClassifier(class_weight="balanced",
                           n_estimators=100,
                           random_state=42),
    "XGBoost":
    XGBClassifier(scale_pos_weight=len(Y_train[Y_train == 0]) /
                  len(Y_train[Y_train == 1]),
                  use_label_encoder=False,
                  eval_metric="logloss",
                  random_state=42),
    "LightGBM":
    LGBMClassifier(class_weight="balanced", random_state=42)
}

# Create a list to store results
results1 = []

# Run each model
for name, model in models.items():
    print(f"\n{'='*30} {name} {'='*30}")

    # Create pipeline with SMOTE and model
    pipe = ImbPipeline([('scaler', StandardScaler()),
                        ('smote', SMOTE(random_state=42)), ('model', model)])

    pipe.fit(X_train, Y_train)
    Y_pred = pipe.predict(X_test)
    Y_prob = pipe.predict_proba(X_test)[:, 1]  # for ROC AUC and PR AUC

    # Evaluation
    print(confusion_matrix(Y_test, Y_pred))
    print('-' * 60)
    print(classification_report(Y_test, Y_pred, digits=4))
    print('-' * 60)
    roc_auc = roc_auc_score(Y_test, Y_prob)
    print("ROC AUC:", roc_auc)
    print('-' * 60)
    pr_auc = average_precision_score(Y_test, Y_prob)
    print("PR AUC:", pr_auc)
    print('-' * 60)
    mcc = matthews_corrcoef(Y_test, Y_pred)
    print("MCC:", mcc)
    print('-' * 60)

    # For the results_df
    # Classification report
    report = classification_report(Y_test, Y_pred, output_dict=True)

    # Extract class 1 metrics (positive class/diverted)
    recall_1 = report["1"]["recall"]
    precision_1 = report["1"]["precision"]
    f1_1 = report["1"]["f1-score"]

    # Append to results list
    results1.append({
        "Model": name,
        "Recall_1": recall_1,
        "Precision_1": precision_1,
        "F1_1": f1_1,
        "ROC_AUC": roc_auc,
        "PR_AUC": pr_auc,
        "MCC": mcc
    })

    # ---------
    # Plot confusion matrix
    sns.heatmap(
        confusion_matrix(Y_test, Y_pred),
        annot=True,
        fmt="d",
        cmap="Blues",
        linewidths=2,  # thickness of the lines between cells
        linecolor='black',  # color of the lines (frame)
        cbar=False  # optional: hide color bar for clarity
    )
    plt.title(f"Confusion Matrix - {name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.tight_layout()
    plt.show()
# ---------
#     # ROC Curve
#     RocCurveDisplay.from_predictions(Y_test, Y_prob)
#     plt.title(f"ROC Curve - {name}")
#     plt.show()

# Convert results to DataFrame
results_df1 = pd.DataFrame(results1)
# save
results_df1.to_pickle(
    "../data/uc5_outputs/predictive_score/predictive_score_GC_trtr_{}.pkl".
    format(num_samples))

# Show results table sorted by PR AUC
# results_df1.sort_values(by="PR_AUC", ascending=False, inplace=True)
print(f"\n{'='*30} {'Final Evaluation Table'} {'='*30}")
print(results_df1.to_string(index=False))
````
